```python
import netCDF4
import matplotlib.pyplot as plt
import os
import shutil

folder_path = "/Users/joonkng/Desktop/pandasApp/"
output_folder = "/Users/joonkng/Desktop/pandasApp/plots/"

os.makedirs(output_folder, exist_ok=True)

for i in range(1, 41):
    file_path = folder_path + f"crop_calendar_{i}.nc4"
    data = netCDF4.Dataset(file_path, "r")
    
    planting_day = data.variables['planting_day'][:]
    lon = data.variables['lon'][:]
    lat = data.variables['lat'][:]
    
    plt.figure()
    plt.imshow(planting_day, origin='upper', extent=[lon.min(), lon.max(), lat.min(), lat.max()])
    plt.colorbar()
    plt.title(f"Planting Day - File {i}")
    plt.xlabel("Longitude")
    plt.ylabel("Latitude")
zip_path = "/Users/joonkng/Desktop/pandasApp/planting_day_maps.zip"
shutil.make_archive(zip_path.replace('.zip', ''), 'zip', output_folder)

print("All images zipped successfully!")
```
